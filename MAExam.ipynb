```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
```

```python
data23 = pd.read_excel('PerformanceMeasurement_2023.xlsx')
data22 = pd.read_excel('PerformanceMeasurement_2022.xlsx')
```

```python
df = pd.concat([data22, data23])
df['PerLost'] = df['NumbLost'] / df['TotalCustomers'] * 100
df['PerNew'] = df['NumbNew'] / df['TotalCustomers'] * 100
#plt.scatter(df.Date, df.NumbLost)
#plt.plot(df.Date, df.NumbNew)
df = df.sort_values(by = "Date")
```

```python
fig = go.Figure()
fig.add_trace(
    go.Line(x = list(df.Date), y = list(df.PerNew))
)

fig.update_layout(title_text = "Plot of new vs lost customers")

fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```

```python
data_melted = df.melt(id_vars=['Date'], value_vars=['PerNew', 'PerLost'],
                                 var_name='Metric', value_name='Ratio')

# Create the plot
fig = px.line(data_melted, x='Date', y='Ratio', color='Metric',
              title='New and Lost Customer Ratios (2022-2023)',
              labels={'Ratio': 'Customer Ratio (%)', 'Date': 'Date'})

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label='1m', step='month', stepmode='backward'),
                dict(count=6, label='6m', step='month', stepmode='backward'),
                dict(count=1, label='1y', step='year', stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(visible=True),
        type='date'
    )
)

# Show the plot
#fig.show()


# Customize the layout
fig.update_layout(
    title='New and Lost Customer Ratios (2022-2023)',
    xaxis_title='Date',
    yaxis_title='Customer Ratio (%)',
    legend_title='Metric',
    hovermode='x unified'
)

# Customize line styles and markers
fig.update_traces(mode='lines+markers')

# Show the customized plot
fig.show()
```
